what is pivot table ?

       A pivot table converts raw data into a matrix format, which is essential for collaborative filtering in recommendation engines.
           
A pivot table reshapes data by selecting:

    Rows (index)
    Columns
    Values to fill the table
    Optional aggregation function (mean, sum, count, etc.)

```python
user_movie_matrix = ratings.pivot_table(index="userId",columns="movieId",values="rating")
```

```python
user_movie_matrix
```

-------  User-Based Collaborative Filtering  --------

```python
user_movie_fill_values = user_movie_matrix.fillna(0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
user_similarity = cosine_similarity(user_movie_fill_values)
```

```python
user_similarity_df = pd.DataFrame(user_similarity,index=user_movie_matrix.index,columns=user_movie_matrix.index)
```

```python
user_similarity_df
```

```python
import pandas as pd

def get_user_recommendations(user_id, n=5):
    # 1. Sabse similar 10 users nikalna
    similar_users = user_similarity_df[user_id].sort_values(ascending=False)[1:11]
    
    # 2. Un similar users ki sabhi movies ki ratings nikalna
    sim_user_ratings = user_movie_fill_values.loc[similar_users.index]
    
    # 3. Weighted Average calculate karo
    weights = similar_users.values.reshape(-1, 1)
    weighted_scores = (sim_user_ratings * weights).sum(axis=0) / weights.sum()
    
    # 4. Jo movies target user ne pehle se dekhi hain, unhe hata dena
    user_seen_movies = user_movie_fill_values.loc[user_id]
    unseen_movies_scores = weighted_scores[user_seen_movies == 0]
    
    # 5. Top 'n' movies select karna (Yahan humne sirf .index nahi liya, poori series rakhi hai)
    top_movies = unseen_movies_scores.sort_values(ascending=False).head(n)
    
    # Data type mismatch error (blank table) se bachne ke liye index ko int mein convert kar dete hain
    top_movies.index = top_movies.index.astype(int)
    
    # 6. 'movies' DataFrame se un ID's ki details nikalna
    recommended_movies = movies[movies['movieId'].isin(top_movies.index)][['movieId', 'title', 'genres']].copy()
    recommended_movies['movieId'] = recommended_movies['movieId'].astype(int)
    
    # 7. PREDICTED RATING ADD KARNA (Map function ka use karke)
    recommended_movies['predicted_rating'] = recommended_movies['movieId'].map(top_movies)
    
    # 8. Decimals ko 2 digits tak limit karna aur highest rating ko upar rakhna
    recommended_movies['predicted_rating'] = recommended_movies['predicted_rating'].round(2)
    recommended_movies = recommended_movies.sort_values(by='predicted_rating', ascending=False).reset_index(drop=True)
    
    return recommended_movies

# Output check karne ke liye run karein:
# print(get_user_recommendations(user_id=1, n=5))
```

```python
def item_based_recommendation_simplest(user_id, n=5):
    # 1. User ne jin movies ko rate kiya hai, unhe nikalo
    user_ratings = user_movie_fill_values.loc[user_id]
    rated_movies = user_ratings[user_ratings > 0]
    
    # Sirf un rated movies ki similarity columns uthao
    similarities = item_similarity_df[rated_movies.index]
    
    # Step B: Seedha '*' symbol se multiply karo
    # Pandas automatically har movie ki rating ko uski similarity se multiply kar dega
    multiplied_scores = similarities * rated_movies
    
    # Step C: Har target movie ka total score nikalo (horizontally plus karke)
    total_scores = multiplied_scores.sum(axis=1)
    
    # Step D: Har target movie ke liye total similarities ka sum nikalo
    sum_of_similarities = similarities.sum(axis=1)
    
    # Step E: Average (Predicted Rating) nikalne ke liye seedha divide (/) karo
    predicted_ratings = total_scores / sum_of_similarities
    # ---------------------------------------------------------
    
    # 3. Jo movies user ne already dekh li hain, unhe hata do
    predicted_ratings = predicted_ratings[user_ratings == 0]
    
    # 4. Top 'n' movies select karna aur index ko int banana
    top_movies = predicted_ratings.nlargest(n)
    top_movies.index = top_movies.index.astype(int)
    
    # 5. Final Recommendation details aur predicted rating add karna
    recommended_movies = movies[movies['movieId'].isin(top_movies.index)][['movieId', 'title', 'genres']].copy()
    recommended_movies['predicted_rating'] = recommended_movies['movieId'].map(top_movies).round(2)
    
    # 6. Sort karke clean table return karna
    recommended_movies = recommended_movies.sort_values(by='predicted_rating', ascending=False).reset_index(drop=True)
    return recommended_movies
```
